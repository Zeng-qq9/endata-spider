### 1.1 实验概述
```python
电影行业近年来得到国家的大力支持，电影行业也在一系列的产业政策支持实现了快速发展，电影的质量和市场反响都得到很大的提升。
从国内的一系列经济数据来看，随着国内生产总值、居民的人均收入不断增长，尤其是对精神文化消费的增长，带动我国电影消费的增长，为我国近年来电影产业发展提供了强大的经济推动力量。从市场上看，我国电影行业的票房也为我国电影行业的发展奠定了基础，2018年我国电影行业的票房收入突破了650亿元。
从 2008年到2018年，中国电影票房以年均35%的速度迅速上升。2017年，中国电影市场更是一跃成为全球第二大电影市场。
2020年受疫情影响，包括电影在内的娱乐行业都受到不小的冲击，为此我们只用到了2010-2019年10年间国内的票房数据，数据展示了当年总票房排名靠前的电影，通过这些数据来具体分析一下这10年国内的电影情况
```

````python
df['新片长'].mean()
````

````python
df['新片长'].max()
````

````python
df['新片长'].min()
````

由上面数据显示，电影片长平均为124分钟左右，最长的达到了三个多小时，而最短的也有90分钟。
```python
其实电影的片长是受一定限制的，尤其是在国内，戏院院商和电影版权持有人大都不赞成拍一部超过100分钟的长片。
因为，电影稍长，每天的放映场次便会减少一场，换言之，即是少了一场电影的营业收益。所以90分钟被认为是最佳长度。 
但是只要故事精彩，时间就不再成为制约导演的最大问题，如194分钟的《泰坦尼克号》，可谓经典中的经典。
现在，长片倒是很普遍的现象了。```

````python
df.head(1)
````

````python
#分析一共有多少种类型
df['新类型'].nunique()
````

````python
##每年各种不同类型的电影的数量分别是多少？
di = df.类型.str.split("|")
di
````

````python
dg = set()
for i in di:
    for x in i:
        dg.add(x)    
dg
````

````python
for type in dg:
    df[type] =df.新类型.str.contains(type).apply(lambda x:1 if x else 0)
````

````python
dk = df[dg]
````

````python
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.commons.utils import JsCode


fn = """
    function(params) {
        if(params.name == '其他')
            return '\\n\\n\\n' + params.name + ' : ' + params.value + '%';
        return params.name + ' : ' + params.value + '%';
    }
    """


def new_label_opts():
    return opts.LabelOpts(formatter=JsCode(fn), position="center")


c = (
    Pie()
     .add(
        "",
        [list(z) for z in zip(["动作", "其他"], [45.3, 54.7])],
        center=["55%", "30%"],
        radius=[60, 80],
        label_opts=new_label_opts(),
    )
    .add(
        "",
        [list(z) for z in zip(["喜剧", "其他"], [6, 94])],
        center=["20%", "30%"],
        radius=[60, 80],
        label_opts=new_label_opts(),
    )
   
    .add(
        "",
        [list(z) for z in zip(["爱情", "其他"], [11, 89])],
        center=["20%", "70%"],
        radius=[60, 80],
        label_opts=new_label_opts(),
    )
   .add(
        "",
        [list(z) for z in zip(["科幻", "其他"], [6.6, 93.4])],
        center=["55%", "70%"],
        radius=[60, 80],
        label_opts=new_label_opts(),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="中国人最喜欢的五种电影类型"),
        legend_opts=opts.LegendOpts(
            type_="scroll", pos_top="20%", pos_left="80%", orient="vertical"
        ),
    )
    .render_notebook()
)
c
````

由多功能饼图我们不难看出，动作片占到了所有影片中的45.3%，明显高于其他电影类型。想到之前分析的这些电影中主演最多的演员是武打片演员成龙，还有范·迪塞尔主演的速度与激情系列，都与动作片挂钩。动作片可谓是国人最爱的电影类型。

````python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.commons.utils import JsCode

x_data = ["2010", "2011", "2012", "2013", "2014", "2016", "2017", "2018"]
y_data1 = [0, 0, 1, 1, 2, 1, 2, 4]
y_data2 = [3, 4, 13, 14, 10, 11, 15, 12]
y_data3 = [1, 1, 2, 3, 6, 3, 1, 3]
y_data4 = [2, 2, 0, 1, 1, 3, 1, 2]
y_data5 = [2, 2, 1, 1, 3, 0, 2, 1]

js_formatter = """function (params) {
        console.log(params);
        return '类型  ' + params.value + (params.seriesData.length ? '：' + params.seriesData[0].data : '');
    }"""

(
    Line(init_opts=opts.InitOpts(width="700px", height="350px"))
    .add_xaxis(
        xaxis_data=x_data
    )
    .extend_axis(
        xaxis_data=x_data,
        xaxis=opts.AxisOpts(
            type_="category",
            axistick_opts=opts.AxisTickOpts(is_align_with_label=True),
            axisline_opts=opts.AxisLineOpts(
                is_on_zero=False, linestyle_opts=opts.LineStyleOpts(color="#6e9ef1")
            ),
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True, label=opts.LabelOpts(formatter=JsCode(js_formatter))
            ),
        ),
    )
    .add_yaxis(
        series_name="喜剧",
        is_smooth=True,
        symbol="emptyCircle",
        is_symbol_show=False,
        # xaxis_index=1,
        color="#d14a61",
        y_axis=y_data1,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=2),
    )
    .add_yaxis(
        series_name="动作",
        is_smooth=True,
        symbol="emptyCircle",
        is_symbol_show=False,
        color="#6e9ef1",
        y_axis=y_data2,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=2),
    )
    .add_yaxis(
        series_name="爱情",
        is_smooth=True,
        symbol="emptyCircle",
        is_symbol_show=False,
        color="#00FA9A",
        y_axis=y_data3,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=2),
    )
    .add_yaxis(
        series_name="奇幻",
        is_smooth=True,
        symbol="emptyCircle",
        is_symbol_show=False,
        color="#808000",
        y_axis=y_data4,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=2),
    )
    .add_yaxis(
        series_name="科幻",
        is_smooth=True,
        symbol="emptyCircle",
        is_symbol_show=False,
        color="#FFFF00",
        y_axis=y_data5,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=2),
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(),
        tooltip_opts=opts.TooltipOpts(trigger="none", axis_pointer_type="cross"),
        xaxis_opts=opts.AxisOpts(
            type_="category",
            axistick_opts=opts.AxisTickOpts(is_align_with_label=True),
            axisline_opts=opts.AxisLineOpts(
                is_on_zero=False, linestyle_opts=opts.LineStyleOpts(color="#d14a61")
            ),
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True, label=opts.LabelOpts(formatter=JsCode(js_formatter))
            ),
        ),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            splitline_opts=opts.SplitLineOpts(
                is_show=True, linestyle_opts=opts.LineStyleOpts(opacity=1)
            ),
        ),
    )
    .render_notebook()
)
````

```python
从年份上来对比类型数量排行前5的类型，可以明显看出，动作片的数量明显高于其他的类型。
而爱情篇的走势和其他类型不同，像一个小山峰，在2014年达到峰值，现在逐年又下降了，不得不说，国内早些年爱情主题的电影还是比较多的，但由于一些原因，比如说编剧，制作效果，审查制度，能排在榜的纯爱情片越来越少。
其他类型基本走势近些年没有变化，国人口味基本没变，还是动作片在国内更受欢迎，这与其震撼效果，演员收益，且类型老少皆宜、受众广都有所关系。```
